## 4. Zaawansowane struktury danych (Series, Data Frame).

**Pandas Series** (tzw. seria danych) jest jednowymiarowym tablicowym obiektem danych, który składa się z jednej kolumny danych i odpowiadających im etykiet (indeksów). Series można porównać do kolumny w arkuszu kalkulacyjnym lub jednowymiarowej tablicy.

Każdy element w Series ma unikalny indeks, który umożliwia dostęp do danych. Indeks może być domyślnie generowany przez Pandas lub można go określić samodzielnie. Series może przechowywać dane różnego typu, takie jak liczby całkowite, zmiennoprzecinkowe, łańcuchy znaków itp.

**Pandas DataFrame** (tzw. ramka danych) jest dwuwymiarową strukturą danych, która składa się z etykietowanych kolumn i indeksowanych wierszy. DataFrame można porównać do tabeli w bazie danych lub arkusza kalkulacyjnego. Każda kolumna w DataFrame jest obiektem typu Series.

DataFrame pozwala przechowywać dane o różnych typach w kolumnach i umożliwia wykonywanie różnych operacji, takich jak filtrowanie, sortowanie, grupowanie i obliczanie statystyk na danych.

Każda ramka danych posiada dwa rodzaje indeksów:
* Indeks kolumnowy (ang. *column label* lub *columns index*) - nazwa/identyfikator kolumny.
* Indeks wierszowy (ang. *row index*) - nazwa/identyfikator wiersza.

Indeksy pozwalają nam łatwo lokalizować poszczególne obserwacje zawarte w ramce danych i przypisane im wartości.

![](zdjęcia/pandas.webp)

### Po co nam Series i DataFrame?

Python w przeciwieństwie do np. języka R nie posiada wbudowanych zaawansowanych struktur danych. Za pomocą Pandas Series i Pandas DataFrame można łatwo wczytywać, manipulować, analizować i wizualizować dane, co czyni Pandas niezwykle przydatnym narzędziem w analizie danych w języku Python.

### Co Pandas wnosi ponad wbudowane struktury danych?

**1. Przetwarzanie danych tabularnych:** Pandas jest zoptymalizowaną biblioteką do manipulacji i analizy danych tabularnych. DataFrame i Series w Pandas są stworzone do efektywnego przetwarzania dużych zbiorów danych w formie tabel. 

**2. Indeksowanie i selekcja danych:** Pandas oferuje zaawansowane możliwości indeksowania i selekcji danych. Możesz dostępować do danych w DataFrame i Series za pomocą etykiet, indeksów numerycznych, zakresów indeksów itp.

**3. Masa dodatkowej funkcjonalności:** Series i DataFrame dostarczają mnóstwo funkcji do filtrowania, sortowania, grupowania, agregacji, łączenia i innych operacji na danych.

**4. Efektywne operacje wektorowe:** Pandas wykorzystuje efektywne operacje wektorowe z biblioteki NumPy, co przyspiesza przetwarzanie danych w porównaniu do tradycyjnych pętli w Pythonie.

**5. Integracja z innymi bibliotekami:** Pandas jest często używany jako część ekosystemu Pythona do analizy danych, wraz z innymi bibliotekami takimi jak NumPy, Matplotlib, scikit-learn itp. Dzięki temu możesz łatwo integrować Pandas z innymi narzędziami do analizy, wizualizacji i modelowania danych.

## 8. Łączenie zbiorów.

Podział metod łączenia danych w Pandas może obejmować:

1. **Concatenate**: Metoda `pd.concat()` w Pandas jest używana do konkatenacji (łączenia) dwóch lub więcej obiektów DataFrame wzdłuż jednej osi (wierszy lub kolumn). Jest to przydatne, gdy chcemy połączyć dane pionowo (wzdłuż osi wierszy) lub poziomo (wzdłuż osi kolumn) bez uwzględniania klucza łączenia.

Merge jest bardziej elastyczny i potężniejszy niż join, ponieważ pozwala na różne rodzaje łączenia (np. lewostronne, prawostronne, wewnętrzne, zewnętrzne) oraz na łączenie na podstawie wielu kluczy. W Pandas istnieje również metoda `join`, ale `merge` i `concatenate` są bardziej wszechstronne i elastyczne. Mogą obsłużyć bardziej złożone operacje łączenia danych, takie jak różne typy łączeń, łączenie na podstawie wielu kluczy czy konkatenacja wielu obiektów danych.

## 10. Grupowanie danych.

### 10.1. Definicja procesu grupowania
Grupowanie jest procesem dzielenia zbioru na grupy obserwacji, na podstawie określonych kryteriów, np. wartości w kolumnie/-ach.

### 10.2. Cel procesu grupowania
Celem grupowania jest zazwyczaj wykonywanie złożonych operacji na podzbiorach danych, które są logicznie powiązane ze sobą (np. klienci tej samej płci, bądź klienci pochodzący z tego samego miasta). Na grupach można wykonywać różne operacje, takie jak obliczanie sumy, średniej, liczby elementów, etc.

1. Utworzone, zagregowane wartości mogą służyć jako charakterystyki, na potrzeby **porównania różnych grup** wnioskowania na ich temat.
2. Innym, częstym celem jest **wzbogacenie zbioru uczącego w modelowaniu predykcyjnym**.

## 12. Próbkowanie.
Próbkowanie (ang. *sampling*) jest procesem wybierania podzbioru obserwacji z większego zbioru danych w celu reprezentacji populacji lub zbioru bazowego. W kontekście modelowania danych, próbkowanie polega na wyborze próby, która stanowi reprezentację większego zbioru danych.

Próbkowanie wykonujemy w kilku względów:
1. **Efektywność obliczeniowa**: Próbkowanie pozwala na zmniejszenie rozmiaru danych, co jest szczególnie przydatne, gdy mamy do czynienia z dużymi zbiorami danych. Próbkowanie umożliwia budowanie modeli na podstawie mniejszej, ale reprezentatywnej próbki danych, co przyspiesza proces modelowania.
2. **Zapobieganie przeuczeniu** (ang. *overfitting*): Próbkowanie jest również używane do zapobiegania przeuczeniu modelu. Przeuczenie występuje, gdy model jest zbyt mocno dopasowany do danych treningowych i nie generalizuje dobrze na nowe dane. Próbkowanie pomaga ograniczyć ten efekt, dostarczając różnorodnych próbek danych do treningu modelu.
3. **Ocena modelu**: Próbkowanie jest często stosowane do oceny jakości modelu. Można podzielić dane na zbiór treningowy i zbiór testowy. Model jest trenowany na zbiorze treningowym, a następnie oceniany na zbiorze testowym, który nie był wykorzystywany w procesie trenowania. Pozwala to ocenić zdolność modelu do generalizacji na nowych danych.

Pandas obsługuje zapisywanie danych do formatu Parquet za pomocą funkcji `to_parquet()` oraz odczytywanie danych z formatu Parquet za pomocą `read_parquet()`.
* **Format Feather**: Feather to inny format zoptymalizowany do szybkiego zapisu i odczytu danych tabelarycznych. Jest to format binarny, który zachowuje typy danych i strukturę, co umożliwia szybkie wczytywanie danych. Pandas obsługuje zapisywanie danych do formatu Feather za pomocą funkcji `to_feather()` oraz odczytywanie danych z formatu Feather za pomocą `read_feather()`.
