## 2. Obsługa braków danych.

### Ćwiczenie 2.1.
Wczytaj plik "churn_banking.csv" do zmiennej `churn_banking` znajdujący się w folderze "dane".

```python
# tu wpisz swój kod
churn_banking = pd.read_csv('dane/churn_banking.csv', index_col='RowNumber')
churn_banking.head()
```

### Ćwiczenie 2.2.
Sprawdź, które kolumny zawierają brakujące wartości i jaka jest ich skala.

```python
# tu wpisz swój kod
churn_banking.isnull().sum()
```

### Ćwiczenie 2.3.
Obsłuż brakujące wartości w kolumnach `Age`, `Balance`, `EstimatedSalary` i `CreditScore` poprzez usunięcie tych obserwacji, które zawierają w nich braki. Nadpisz zmienną `churn_banking`.

ps. Zwróć uwagę na atrybut `subset` metody `dropna`.

```python
churn_banking = churn_banking.dropna(subset=['Age', 'Balance', 'EstimatedSalary', 'CreditScore'])
```

```python
churn_banking.isnull().sum()
```

### Ćwiczenie 2.4.
Braki zawiera też kolumna `Gender`. Być może ktoś nie chciał podawać swojej płci? A może nie czuje swojej przynależności do żadnej z nich? Zamiast je usuwać wypełnijmy je wartością "Other". Nadpisz zmienną `churn_banking`.

```python
# tu wpisz swój kod
churn_banking.fillna('Other', inplace=True)
churn_banking.head()
```

```python
churn_banking.isnull().sum() # brak braków ;-)
```

```python
churn_banking.Gender.value_counts() # istnieje nowa kategoria w zmiennej "Gender"
```

```python
# tu wpisz swój kod
wages.earn.mean()
```

### Ćwiczenie 5.2.
Pogrupuj teraz ten sam plik "wages.txt" po zmiennej "sex". Dla każdej z płci wyznacz średnią wartość zarobków. Czy widzisz tu jakieś różnice pomiędzy grupami? Czy mężczyźni zarabiają średnio więcej niż kobiety?

```python
# tu wpisz swój kod
wages.groupby('sex').earn.mean()
# wygląda na to, że mężczyźni średnio zarabiają znacznie więcej niż kobiety
```

### Ćwiczenie 5.3.
W poprzednić ćwiczeniu widoczna jest pewna różnica w grupach. Czy ta różnica jest widoczna dla wszystkich ras? Wykonaj ponownie grupowanie, lecz teraz użyj dwóch zmiennych, by odpowiedzieć na powyższe pytanie. 

```python
# tu wpisz swój kod
wages.groupby(['race', 'sex']).earn.mean()

# różnica w zarobkach pomiędzy kobietami a mężczyznami jest różna w zależności od rasy
    # największa dysproporcja występuje u rasy białej - "white" - tu mężczyźni zarabiają średnio znacznie więcej niż kobiety
    # w rasach "hispanic" i "black" różnica w zarobkach pomiędzy kobietami i mężczyznami jest mniejsza niż w przypadku rasy "white"
    # co ciekawe w kategorii "other" ten trend jest zupełnie odwrócony i to kobiety zarabiają średnio węcej niż mężczyźni
```

## 6. Próbkowanie.

### Ćwiczenie 6.1.
Wylosuj ze zwracaniem próbę 1000 obserwacji z pliku "wages.txt" i sprawdź średnią wielkośc zarobków dla próby.
* Zadbaj o replikowalnośc wyników.
* Czy średnia różni się od średniej dla oryginalnej próby?
* Jak duża jest różnica?

```python
# tu wpisz swój kod
wages_sample = wages.sample(1000, replace=True, random_state=2001)
print('Średnia zarobków w całej próbie:', wages.earn.mean().round(2))
print('Średnia zarobków w próbie losowanej:', wages_sample.earn.mean().round(2))
```

### Ćwiczenie 6.2.
Korzystając z funkcji `train_test_split` znajdującej się w module `sklearn.model_selection` podziel zbiór "wages.txt" na część uczącą i testową w proporcji 3:1 i zapisz nowe zbiory do zmiennych: `wages_tr`, `wages_te`.

Sprawdź czy każdy nowo powstały zbiór ma odpowiednią wielkość za pomoca właściwości zbioru - `.shape`.

### Ćwiczenie 6.3.
